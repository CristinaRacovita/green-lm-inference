### 0. Import libraries

```python
import pandas as pd
```

### 1. Load and prepare measurements

```python
data_path = "../results/rq3_embedding_generation/"
measurements_name = "measurements_gte_small_nfcorpus.csv"
timestamps_name = "timestamps_gte_small_nfcorpus.csv"
```

```python
measurements = pd.read_csv(data_path + measurements_name, encoding="unicode_escape")[:-2]
print(f"The shape of the measurements dataset is: {measurements.shape}")
measurements.head()
```

```python
# create a data-time column
date_time = zip(measurements["Date"].to_list(), measurements["Time"].to_list())
measurements["timestamp"] = [date_val + " " + time_val for date_val, time_val in date_time]
measurements["timestamp"] = pd.to_datetime(measurements["timestamp"], format="%d.%m.%Y %H:%M:%S.%f")
```

```python
# remove the columns that are not used
columns_to_drop = ["Date", "Time"]
measurements = measurements.drop(columns_to_drop, axis=1)
measurements.head()
```

### 2. Load and prepare timestamps of experiment runs

```python
run_timestamps = pd.read_csv(data_path + timestamps_name, index_col="run_number")
run_timestamps
```

```python
start_timestamps = pd.to_datetime(run_timestamps["start_timestamp"]).to_list()
end_timestamps = pd.to_datetime(run_timestamps["end_timestamp"]).to_list()
```

### 3. Analyze runs

```python
# compute for each experiment run the duration and get the associated measurements
timestamps = measurements["timestamp"].to_numpy()
experiment_intervals = zip(start_timestamps, end_timestamps)

for interval in experiment_intervals:
    duration_seconds = (interval[1] - interval[0]).seconds
    duration_microseconds = (interval[1] - interval[0]).microseconds
    duration = duration_seconds + duration_microseconds / 1000000

    considered_timestamps = timestamps[(interval[0] <= timestamps) & (timestamps <= interval[1])]
    experiment_measurements = measurements[measurements["timestamp"].isin(considered_timestamps)]
```
